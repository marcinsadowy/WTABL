```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

z_h = 100
z_0hi = 0.92
u_star = 0.29
k = 0.4
z_0lo = 0.0001
beta = 0.68
D = 100
delta_ibl_0 = z_h
```

```python
x = np.linspace(0, 5000, 5000)

delta_ibl_x = delta_ibl_0 + z_0hi * (x/z_0hi)**(4/5)
U_hT_x = (u_star / k) * (np.log(delta_ibl_x / z_0lo) / np.log(delta_ibl_x / z_0hi)) * np.log((z_h / z_0hi) * (1 + D/(2 * z_h))**beta)

array_2d = np.zeros((5000, 5000))

for i in range(array_2d.shape[1]):
    array_2d[:, i] = U_hT_x[i]

def add_line(array, start_x, start_y, length_x, length_y, value):
    array[start_y:start_y + length_y, start_x:start_x + length_x] = value

# Define the starting points and lengths of the lines
lines = [
    (1000, 1000, 5, 100),
    (1000, 2000, 5, 100),
    (1000, 3000, 5, 100),
    (1000, 4000, 5, 100),
    (1000, 5000, 5, 100),
    (1000, 6000, 5, 100),
    (1000, 7000, 5, 100),
    (1000, 8000, 5, 100),
    (1000, 9000, 5, 100),
    (1000, 10000, 5, 100),
    (2000, 1000, 5, 100),
    (2000, 2000, 5, 100),
    (2000, 3000, 5, 100),
    (2000, 4000, 5, 100),
    (2000, 5000, 5, 100),
    (2000, 6000, 5, 100),
    (2000, 7000, 5, 100),
    (2000, 8000, 5, 100),
    (2000, 9000, 5, 100),
    (2000, 10000, 5, 100),
    (3000, 1000, 5, 100),
    (3000, 2000, 5, 100),
    (3000, 3000, 5, 100),
    (3000, 4000, 5, 100),
    (3000, 5000, 5, 100),
    (3000, 6000, 5, 100),
    (3000, 7000, 5, 100),
    (3000, 8000, 5, 100),
    (3000, 9000, 5, 100),
    (3000, 10000, 5, 100),
    (4000, 1000, 5, 100),
    (4000, 2000, 5, 100),
    (4000, 3000, 5, 100),
    (4000, 4000, 5, 100),
    (4000, 5000, 5, 100),
    (4000, 6000, 5, 100),
    (4000, 7000, 5, 100),
    (4000, 8000, 5, 100),
    (4000, 9000, 5, 100),
    (4000, 10000, 5, 100),
    (5000, 1000, 5, 100),
    (5000, 2000, 5, 100),
    (5000, 3000, 5, 100),
    (5000, 4000, 5, 100),
    (5000, 5000, 5, 100),
    (5000, 6000, 5, 100),
    (5000, 7000, 5, 100),
    (5000, 8000, 5, 100),
    (5000, 9000, 5, 100),
    (5000, 10000, 5, 100),
    (6000, 1000, 5, 100),
    (6000, 2000, 5, 100),
    (6000, 3000, 5, 100),
    (6000, 4000, 5, 100),
    (6000, 5000, 5, 100),
    (6000, 6000, 5, 100),
    (6000, 7000, 5, 100),
    (6000, 8000, 5, 100),
    (6000, 9000, 5, 100),
    (7000, 10000, 5, 100),
    (7000, 1000, 5, 100),
    (7000, 2000, 5, 100),
    (7000, 3000, 5, 100),
    (7000, 4000, 5, 100),
    (7000, 5000, 5, 100),
    (7000, 6000, 5, 100),
    (7000, 7000, 5, 100),
    (7000, 8000, 5, 100),
    (7000, 9000, 5, 100),
    (7000, 10000, 5, 100),
    (8000, 1000, 5, 100),
    (8000, 2000, 5, 100),
    (8000, 3000, 5, 100),
    (8000, 4000, 5, 100),
    (8000, 5000, 5, 100),
    (8000, 6000, 5, 100),
    (8000, 7000, 5, 100),
    (8000, 8000, 5, 100),
    (8000, 9000, 5, 100),
    (8000, 10000, 5, 100),
    (9000, 1000, 5, 100),
    (9000, 2000, 5, 100),
    (9000, 3000, 5, 100),
    (9000, 4000, 5, 100),
    (9000, 5000, 5, 100),
    (9000, 6000, 5, 100),
    (9000, 7000, 5, 100),
    (9000, 8000, 5, 100),
    (9000, 9000, 5, 100),
    (9000, 10000, 5, 100),
    (10000, 1000, 5, 100),
    (10000, 2000, 5, 100),
    (10000, 3000, 5, 100),
    (10000, 4000, 5, 100),
    (10000, 5000, 5, 100),
    (10000, 6000, 5, 100),
    (10000, 7000, 5, 100),
    (10000, 8000, 5, 100),
    (10000, 9000, 5, 100),
    (10000, 10000, 5, 100),
]

# Draw the lines on the array
for line in lines:
    add_line(array_2d, *line, np.max(U_hT_x))

# Create a figure and axis
fig, ax = plt.subplots()
# Display the image
cax = ax.imshow(array_2d, extent=[0, 5000, 0, 5000], aspect='auto', origin='lower')

# Add rectangles for each line
for line in lines:
    rect = patches.Rectangle((line[0], line[1]), line[2], line[3], linewidth=1, edgecolor='w', facecolor='w')
    ax.add_patch(rect)

# Set labels and title
plt.imshow(array_2d, extent=[0, 5000, 0, 5000], aspect='auto', origin='lower', cmap='coolwarm')
plt.colorbar(label='U_hT [m/s]')
plt.xlabel('Downstream distance [m]')
plt.xlim(-2000, 7000)
plt.ylabel('Transverse distance [m]')
plt.ylim(-2000, 7000)
plt.title('2D Array Plot')
plt.show()
```
